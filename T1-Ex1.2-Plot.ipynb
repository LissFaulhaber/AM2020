```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt #biblioteca para plotagem
```

```python
def plot(filepath, theta):
    #prepara leitura dos dados do arquivo
    path = os.getcwd() + filepath
    dataset = pd.read_csv(path, header=None)
    X = dataset.iloc[:, 0:-1].values
    y = dataset.iloc[:, -1:].values
    
    #prepara montagem do gráfico
    #determina o valor mínimo e máximo que a reta "t", saltando de 1 em 1
    t = np.arange(0, 25, 1)
    #define as características dos marcadores de x/y
    plt.scatter(X, y, color='red', marker='x', label='Training Data')
    #define as características da linha de regressão
    plt.plot(t, theta[0] + (theta[1]*t), color='blue', label='Linear Regression')
    #define limites, títulos e rótulos do gráfico
    plt.axis([4, 25, -5, 25])
    plt.title('Populacao da cidade x Lucro da filial')
    plt.xlabel('Populacao da cidade (10k)')
    plt.ylabel('Lucro (10k)')
    plt.legend()
    #nome a ser adotado pela imagem gerada
    filename = 'target/plot1.2.png'
    if not os.path.exists(os.path.dirname(filename)):
        os.makedirs(os.path.dirname(filename))
    #gerar imagem e salvar arquivo
    plt.savefig(filename)
    plt.show()
```

```python
#chama a função criada indicando o caminho do arquivo e os valores de Theta estabelecidos pela equação de Gradiente Descente
# valor de Theta extraído de 'T1-Ex1.2-GD.ipynb'
plot(filepath='/data/ex1data1.txt', theta=[[-3.89530051],[1.19298539]])
```
